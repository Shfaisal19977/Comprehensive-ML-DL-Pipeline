```python
import tensorflow as tf

# Data Loading and Preprocessing Class
class DataLoader:
    def __init__(self, file_path, target_col):
        self.file_path = file_path
        self.target_col = target_col

    def load_data(self):
        dataset = tf.data.experimental.make_csv_dataset(
            self.file_path,
            batch_size=32,
            label_name=self.target_col,
            na_value="?",
            num_epochs=1,
            ignore_errors=True
        )
        return dataset

# Usage Example Function
def run_data_loading(file_path, target_col):
    data_loader = DataLoader(file_path=file_path, target_col=target_col)
    dataset = data_loader.load_data()

    for batch in dataset.take(1):
        features, labels = batch
        print("Features:", features)
        print("Labels:", labels)

# Sample usage with custom dataset
# run_data_loading('path_to_dataset.csv', 'target_column')
```
